```python
import pandas as pd
df = pd.read_csv("/Users/richa/Desktop/python/machine_learning/datasource/carprices.csv")
df
```

```python
# 1. Through Dummy Val
```

```python
# Above we can see 'Car Model' column contains String value, we have to convert this column into dummy val
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
# We can see it has created dummy column for the dataframe
# We have to concat both the dataframe for further analysis
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
# Now we need drop the original town column because we have already created dummy column for the 'Car Model' column
# we have to drop one of the column dummy variable for avoid numpy variables trap, read more about dummy variable trap
# from google
final_df=merged.drop(['Car Model','Mercedez Benz C class'], axis='columns')
final_df
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
```

```python
# here our X is four columns of final_df dataFrame ('Mileage', 'Age(yrs)', 'Audi A5','BMW X5')
# So, we have to create X dataset
X = final_df.drop("Sell Price($)", axis='columns')
X
```

```python
y = df['Sell Price($)']
y
```

```python
# to train the machine from our dataSet
lg.fit(X,y)
```

```python
lg.predict([[45000,4,0,0]])
```

```python
lg.predict([[86000,7,0,1]])
```

```python
lg.score(X,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_le=df
df_le
```

```python
# le.fit_transform(df_le.town) take column name in which label to be given, it will returns array
le.fit_transform(df_le['Car Model'])
```

```python
# We donot want array, we want only label in our dataFrame for the column 'Car Model'
df_le['Car Model'] = le.fit_transform(df['Car Model'])
df_le
```

```python
# We will create our X variable, X  is our traning dataset and y is dependent varaible, we want X in the form 
# 2-D array in One Hot Encoding
X = df_le[['Car Model','Sell Price($)', 'Age(yrs)']].values
X
```

```python
y = df_le['Mileage']
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer #ColumnTransformer
ct = ColumnTransformer([('Car Model', OneHotEncoder(), [0])], remainder='passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
# to remove a column from the X component of the dataset, to prevent dummy trap
X = X[:,1:]
X
```

```python
lg.fit(X,y)
```

```python
lg.predict([[0,1,45000,4]])
```

```python
lg.predict([[0,0,86000,7]])
```

```python
lg.score(X,y)
```
