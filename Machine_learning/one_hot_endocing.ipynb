```python
# Dummy Variables & One Hot Encoding
# Dummy Variables
# Computer machine can only work with numeric data. If we have string data in input,
# first we have convert it in numeric value
import pandas as pd
import numpy as np
df = pd.read_csv('/Users/richa/Desktop/python/machine_learning/datasource/homeprices_with_String_column.csv')
df
```

```python
# From the above file we can see town column has towns name in String we cannot change it to numberic value 
# like 1,2,3.. for each city because during modelling system will proritise the town according to number assigned
# between them such as monroe township > west windsor but this is not true
# Hences we define 'One hot Encoding' in which each town defined as a dummy column as shown below
dummies = pd.get_dummies(df.town)
dummies
```

```python
# We can see it has created dummy column for the dataframe
# We have to concat both the dataframe for further analysis
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
# Now we need drop the original town column because we have already created dummy column for the 'town' column
# we have to drop one of the column dummy variable for avoid numpy variables trap, read more about dummy variable trap
# from google
final_df=merged.drop(['town','monroe township'], axis='columns')
final_df
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# here our x is three columns of final_df dataSet (area, robinsville, west windosor)
# So, we have to create x dataset
x= final_df.drop("price", axis='columns')
x
```

```python
# and y is only the 'price' column of final_df dataSet
y = final_df.price
y
```

```python
# to train the machine from our dataSet
model.fit(x,y)
```

```python
#model.predict(np.array(3300).reshape(-1, 1))
model.predict(np.array([[2800,0,1]]))
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(x,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_le = df
df_le
```

```python
# le.fit_transform(df_le.town) take column name in which label to be given, it will returns array
le.fit_transform(df_le.town)
```

```python
# We donot want array, we want only label in our dataFrame
df_le.town = le.fit_transform(df_le.town)
df_le
```

```python
# We will create our X variable, X  is our traning dataset and y is dependent varaible, we want X in the form of array
# in One Hot Encoding
X = df_le[['town','area']].values
X
```

```python
y = df_le.price
y
```

```python
#ColumnTransformer helps to set 'town' column as encodded column in the 'X' component of the dataset
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict(np.array([[0,1,3400]])) # 3400 sqr ft home in west windsor
```
