```python
import pandas as pd
india_weather= pd.DataFrame({
    'City' : ['Mumbai','New Delhi','Banglore'],
    'Temperature' : [32,35,30],
    'Humidity' : [76,80,66]
})
india_weather
```

```python
us_weather= pd.DataFrame({
    'City' : ['New york','Chicago','Orlando'],
    'Temperature' : [30,20,35],
    'Humidity' : [80,70,85]
})
us_weather
```

```python
df=pd.concat([india_weather,us_weather])
df
```

```python
# to get the correct index
df=pd.concat([india_weather,us_weather], ignore_index=True)
df
```

```python
df=pd.concat([india_weather,us_weather], keys=['india','us'])
df
# with the help of key, we can find both the DataFrame uniquely
```

```python
df.loc["india"]
```

```python
# to append DataFrame column wise
temperature_df=pd.DataFrame({
    'cities':['Mumbai','Banglore','Chennai'],
    'temperature' :[30,32,33]
})
temperature_df
```

```python
windSpeed_df=pd.DataFrame({
    'cities':['Mumbai','Banglore','Chennai'],
    'windspeed' :[6,7,9]
})
windSpeed_df
```

```python
df=pd.concat([temperature_df,windSpeed_df])#axis=0(here)
df
# it is appending data row wise but we need to append our data column wise  by using axis method
```

```python
df=pd.concat([temperature_df,windSpeed_df], axis=1)
df
```

```python
windSpeed_df=pd.DataFrame({
    'cities':['Chennai','Banglore'],
    'windspeed' :[6,9]
})
windSpeed_df
```

```python
df=pd.concat([temperature_df,windSpeed_df], axis=1)
df
# here merging is wrong mumbai is merged with chennai, for solving this we have to put index in DataFrame Creation
```

```python
temperature_df=pd.DataFrame({
    'cities':['Mumbai','Banglore','Chennai'],
    'temperature' :[30,32,33]
}, index=[0,1,2])
temperature_df
```

```python
windSpeed_df=pd.DataFrame({
    'cities':['Chennai','Banglore'],
    'windspeed' :[6,9]
}, index=[2,1])
windSpeed_df
```

```python
df=pd.concat([temperature_df,windSpeed_df], axis=1)
df
```
