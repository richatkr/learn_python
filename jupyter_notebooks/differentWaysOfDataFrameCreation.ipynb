```python
# to create dataframe using dictionary 

weather_data = {
    
    'day' : ['01/01/2017','01/02/2017','01/03/2017'],
    'tempreture' : [30,26,20],
    'windspeed' : [6,7,5],
    'event' : ['Sunny','Rain','Snow']
}
df=pd.DataFrame(weather_data)
df
```

```python
# to create dataframe using List of tuples

weather_data = [
    
    ('01/01/2017',30,6,'Sunny'),
    ('01/02/2017',26,8,'Rain'),
    ('01/03/2017',20,5,'Snow')
]

df=pd.DataFrame(weather_data)
#specifically mention the column during dataframe creation
df=pd.DataFrame(weather_data,columns=["Days","Temperature","WindSpeed","Event"])
df
```

```python
# to create dataframe using List of dictionaries
weather_data = [
    {"days":"01/01/2017", "temperature":30, "windspeed":6, "event":"Sunny"},
    {"days":"01/02/2017", "temperature":20, "windspeed":8, "event":"Rain"},
    {"days":"01/03/2017", "temperature":15, "windspeed":4, "event":"Snow"}

]

df=pd.DataFrame(weather_data)
df

```
