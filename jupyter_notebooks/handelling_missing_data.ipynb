```python
# to make day column as Date we have to specify it in data frame creation
df=pd.read_csv("/Users/richa/Desktop/python/jupyter_notebooks/datasource/weather_data_Newyork.csv", parse_dates=["day"])
df
```

```python
# To make day column as Index
df.set_index('day', inplace=True)
df
```

```python
# If we want to replace all NaN value with some meaningful value, suppose if we want to replace 'na'
# First method is fillna, we will not modify our original dataframe intread we will create new dataframe

new_df=df.fillna(0)
new_df


```

```python
# Above 'event' column value 'NaN' replaced with 0, which doesnot mean any think, to avoid such scenario we can pass
# different values for different column through dictionary
new_df= df.fillna({
    
    'temperature' : 0,
    'windspeed'  : 0,
    'event' : 'No Event'
})
new_df
```

```python
# Here we can see in 'temperature' column it has 0 value, which means temperature was 0 no that day, but this is not
# true, we donot know the temperature on that day. In this situation we use 'Forward Fill'(ffill) 
#and 'Backward Fill'(bfill) method
# 'Forward Fill' fill method take the upper column value and replace the 'NaN or 0' value with upper column value
new_df=df.fillna(method="ffill")
new_df
```

```python
# 'backward Fill' method take the lower column value and replace the 'NaN or 0' value with lower column value
new_df=df.fillna(method="bfill")
new_df
# check in column tempreture , windspeed and event
```

```python
# axis method axis="columns" will work row wise it will take data from event column and update it in windspeed column
# if it has '0 or NaN' value
new_df=df.fillna(method="bfill", axis="columns")
new_df
```

```python
df
```

```python
# limit method , If we want to copy the value in only one row not all subsequent row then we use limit

new_df=df.fillna(method='ffill', limit=1)
new_df
# we can for the row of 8th and 9th jan
```

```python

new_df=df.fillna(method='ffill', limit=2)
new_df
```

```python
# We are still not happy with filled data because 1st jan temperature is 32 and 5th jan temperature is 28 
# 4th Jan Temperature may be in between 32 and 28 , for achieving this we have interpolate method in pandas
new_df=df.interpolate()
new_df
# interpolate(), here we havn't pass any value so it is calculating the average value in the NaN column value
```

```python
new_df=df.interpolate(method='time')
new_df
#df.interpolate(method='time'), it will consider the 4th jan date for which column value to be filled was near to 
#5th jan so average value calcule according to that
```

```python
# if we want to return the rows which donot contains any NaN value
new_df=df.dropna()
new_df
```

```python
# If you want to avoid the row in which all column has NaN value

new_df=df.dropna(how="all")
new_df
```

```python
# If we want to keep the row which at least has one column with non NaN value 
#then we use thresh parameter in dropna method

new_df=df.dropna(thresh=1)
new_df
```

```python
#thresh=2 will keep rows which has at least two non 'NaN' value column
new_df=df.dropna(thresh=2)
new_df
```
