```python
import matplotlib.pyplot as plt
%matplotlib inline
x=[1,2,3,4,5,6,7]
y=[28,20,32,36,34,28,34]
plt.plot(x,y)
```

```python
# other argument of the plot function
plt.plot(x,y, color='green')
```

```python
# other argument of the plot function linewidth for thicker line
plt.plot(x,y, color='green', linewidth=5)
```

```python
# plot function linestyle property helps to represent dotted, dashed, dashdot etc lines in a graph line
plt.plot(x,y, color='green', linewidth=5, linestyle='dotted')
```

```python
# to label the x and y axis and title of the graph we have below function
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, color='green', linewidth=5, linestyle='dotted')
```

```python
# we can also provide design argument by providing string value
plt.plot(x,y, 'gd--')
# means the g is the green color, d is diamond and -- for the dotted lines of the graph,
# we can write this in any sequence i.e 'dg--' or '--gd' is also correct
# 'gd--' is known as String argument
```

```python
plt.plot(x,y,'r*')
```

```python
# we can also write the above as individual argument form
plt.plot(x,y, color='red', marker='D', linestyle='')
```

```python
# we can also write the above as individual argument form and color in hexadecimal form
plt.plot(x,y, color='#008888', marker='D', linestyle='dotted')
```

```python
# to control marker size in the graph
plt.plot(x,y, color='#008888', marker='D', linestyle='dotted', markersize=20)
```

```python
# to control marker size in the graph
plt.plot(x,y, color='#008888', marker='*', linestyle='dashdot', markersize=10)
```

```python
# alpha property , controls the transparency of the graph 0 indicates white , 1 indicates full color
# below is with 0.4 transparency
plt.plot(x,y, color='#008888', marker='*', linestyle='dashdot', markersize=10, alpha=0.4)
```
