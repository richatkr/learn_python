```python
#weather_data_replace_method.csv
import pandas as pd
import numpy as np
df=pd.read_csv("/Users/richa/Desktop/python/jupyter_notebooks/datasource/weather_data_replace_method.csv")
df
```

```python
# to replace -99999 into NaN value
import numpy as np
new_df=df.replace(-99999, np.NaN)
new_df
```

```python
# if we have two garbage value like -99999 and -88888, then we will provide a list
new_df=df.replace([-99999,-88888],np.NaN)
new_df
```

```python
# We should not replace event column 0 value using 'list' because 'replace' method checks every column 
# and replace the given value 'NaN' and 0 for the column windspeed is correct value. So we use dictionary here

new_df=df.replace({
    "temperature" : -99999,
    "windspeed" : -99999,
    "event" : "0"
}, np.NaN)

new_df
```

```python
# We can pass dictionary one other way also, suppose we want to replace -99999 with NaN and 
# 'event' column 0 value with 'Sunny', which means if no data is given we will assume the weather is Sunny

new_df=df.replace({
    -99999 : np.NaN,
    '0' : 'Sunny'
})
new_df
```

```python
# If we have values in 'temperature' column as 32 C or 28 F and 'windspeed' column as 6.0 mph or 7.0 mph
# and want only numeric values in these columns, we can achieve this through regular expression regex method
new_df=df.replace('[A-Za-z]','', regex=True)
new_df
```

```python
# We can see above after replacing event column's all alphabet disappeared, 
#So, we have to use dictionary for replacement

new_df=df.replace({
    'temperature' : '[A-Za-z]',
    'windspeed' : '[A-Za-z]'
},'', regex=True)

new_df
# Now we can see below it is replacing only temperature and windspeed column alphabet values
```

```python
# To replace list of values with another list of values
df=pd.DataFrame({
    'Student':['Navya','Nutan','Navin','Nishant','Nisha','Nirabh','Nirala'],
    'Score' : ['Exceptional','Great','Average','Good','Exceptional','Poor','Great']
    })

df
```
