```python
# Histograms
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[60,110,85,90,75,149,120,75,105,77,80,130,65,90]
plt.hist(blood_sugar)
# In histogram Y-axis contains the frequency of the number of patient in the array list
# By default histogram creates 10 bins
```

```python
# bins argument properties of histogram, by default bin was 10, we have changed it to 3, we can see, 
# we have three bars, but these bars are joined, so we can make it little better by supplying rwidth parameter
# rwidth maximum value which is default value is 1, we will provide here 0.95 to see the changes
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
# Above bin is equally dividing the graph, we can pass the the bin parameter explicity for different sugar level
plt.hist(blood_sugar, bins=[60,100,125,150], rwidth=0.95)
```

```python
# Other properties of hist function color, histtype
plt.hist(blood_sugar, bins=[60,100,125,150], rwidth=0.95 , color='g', histtype='step')
```

```python
# Other properties of hist function color, histtype
plt.hist(blood_sugar, bins=[60,100,125,150], rwidth=0.95 , color='g')
```

```python
# We have two list of sugar level of men and women as below
sugar_level_men=[60,65,110,75,90,120,150,70,90,160,80,85,99]
sugar_level_women=[70,55,149,75,80,88,66,72,110,77,89,66,88]
plt.xlabel('Sugar Level')
plt.ylabel('Total No. of patient')
plt.title('Blood Sugar Analysis')
plt.hist([sugar_level_men,sugar_level_women], bins=[60,100,125,150], rwidth=0.95 , color=['green','orange'],
        label=['Men','Women'])
plt.legend()
```

```python
# To plot histogram horizontally we have orientation property
plt.xlabel('Sugar Level')
plt.ylabel('Total No. of patient')
plt.title('Blood Sugar Analysis')
plt.hist([sugar_level_men,sugar_level_women], bins=[60,100,125,150], rwidth=0.95 , color=['green','orange'],
        label=['Men','Women'],orientation='horizontal')
plt.legend()
```
