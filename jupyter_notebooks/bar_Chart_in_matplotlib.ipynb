```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGLE','FACEBOOK','YOUTUBE','APPLE']
revenue=[100,50,80,60]
profit=[40,20,30,20]
```

```python
# to convert company list into number np.arange(len(company)) , here company is a list of string
xpos=np.arange(len(company))
xpos
```

```python
# to convert ypos again into company name we have function xticks
plt.xticks(xpos,company)
plt.ylabel('Revenue(bln)')
plt.bar(xpos,revenue)
#plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue(bln)')
plt.bar(xpos,revenue, label='Revenue')
plt.legend()
# In finance people plot side by side revenue and profit
```

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue(bln)')
plt.bar(xpos-0.2,revenue, label='Revenue')# this is the reason we have changed the company list to number so that
# we can arrange the bar chart side by side by changing the position of the element
plt.bar(xpos+0.2,profit, label='Profit')

plt.legend()
# This generates the both in same graph but we need side by side graph 
```

```python
# We can see in above graph because of width of graph, it is not showing side by side properly, we can achieve this by
# using width
plt.xticks(xpos,company)
plt.ylabel('Revenue(bln)')
plt.bar(xpos-0.2,revenue, label='Revenue', width=0.4)
plt.bar(xpos+0.2,profit, label='Profit', width=0.4)
plt.legend()
```

```python
# Sometimes we have requirement of creating horizontal bar chart for that we have function called barh
plt.yticks(xpos,company)
#plt.ylabel('Revenue(bln)')
plt.barh(xpos-0.2,revenue, label='Revenue')
plt.barh(xpos+0.2,profit, label='Profit')
plt.legend()
```
