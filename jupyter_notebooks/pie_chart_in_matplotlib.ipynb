```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals=[7000,4000,1200,800,2000]
exp_labels=['House Rent','Food','Travelling','Phone/Internet/cabel Bills','Others']
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
# Other Properties of Pie Chart
plt.pie(exp_vals,labels=exp_labels)
plt.axis('equal')
# plt.show() removes the extra information from the graph
plt.show()
```

```python
# To increase radius of pie char we have radius argument, default value of  radius argument is 1
plt.pie(exp_vals,labels=exp_labels, radius=2)
plt.axis('equal')
plt.show()
```

```python
# To get each expence presentage we have argument autopct='%f%%'
plt.pie(exp_vals,labels=exp_labels, radius=2 , autopct='%f%%')
plt.axis('equal')
plt.show()
```

```python
# To get 2 digit only after decimal autopct='%0.2f%%'
plt.pie(exp_vals,labels=exp_labels, radius=2 , autopct='%0.2f%%')
plt.axis('equal')
plt.show()
```

```python
# If we donot want any decimal value then autopct='%0.0f%%', to get shawdow in pie chart set shadow = True
plt.pie(exp_vals,labels=exp_labels, radius=2 , autopct='%0.0f%%', shadow=True )
plt.axis('equal')
plt.show()
```

```python
# To get one slice out from the pie chart, we have a property explode
# In explode we provide number of values, here no of values= number of pie we have 
# explode=[0,0,0,0,0] nothing will change

plt.pie(exp_vals,labels=exp_labels, radius=2 , autopct='%0.0f%%', shadow=True , explode=[0,0,0,0,0])
plt.axis('equal')
plt.show()
```

```python
# explode=[0,0.1,0,0,0] nothing will change , food pie will come out from the chart

plt.pie(exp_vals,labels=exp_labels, radius=2 , autopct='%0.0f%%', shadow=True , explode=[0,0.1,0,0,0])
plt.axis('equal')
plt.show()
```

```python

plt.pie(exp_vals,labels=exp_labels, radius=2 , autopct='%0.0f%%', shadow=True , explode=[0,0.1,0.5,0,0])
plt.axis('equal')
plt.show()
```

```python
# By default pie chart start with the angle left round anti clock wise, we can change the startangle of pie chart

plt.pie(exp_vals,labels=exp_labels, radius=2 , autopct='%0.0f%%', shadow=True , explode=[0,0.1,0,0,0], startangle=45)
plt.axis('equal')
plt.show()
```
