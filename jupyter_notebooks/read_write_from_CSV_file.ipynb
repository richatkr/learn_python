```python
# We can check here in our file some column has values like 'n.a' or 'not available' , for replacing all 'n.a' 
#or 'not avaliable' into NaN value for cleaning our data we use na_values function syntax is shown below

df=pd.read_csv('/Users/richa/Desktop/python/jupyter_notebooks/datasource/stock_data.csv', 
               na_values=['n.a.','not available'])
df
```

```python
# We can see above revenue column has '-1' which is not possible if we have clean the data and change -1 to NaN value
# if we user na_values=['n.a.','not available','-1'] , this will change 'eps' column -1 value to NaN, but eps column 
# can contain -1 value so intread of using list in na_values we shoould use dictionary as shown below

df=pd.read_csv('/Users/richa/Desktop/python/jupyter_notebooks/datasource/stock_data.csv', 
               na_values={
                 'eps' : ['n.a.','not available'],
                   'revenue' : ['n.a.','not available',-1],
                   'price' : ['n.a.','not available'],
                   'people' : ['n.a.','not available']
                   
               }
              )
df
```

```python
# Write back to the csv file, we have a data frame df we will write back this data frame to new csv file as shown below

df.to_csv('/Users/richa/Desktop/python/jupyter_notebooks/datasource/write_stock_data.csv')
```
