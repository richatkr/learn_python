```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
day=[1,2,3,4,5,6,7]
max_t=[33,34,35,32,33,36,34]
min_t=[28,29,30,27,30,22,20]
avg_t=[30,32,31,29,31,28,22]
```

```python
plt.plot(day,max_t)
plt.plot(day,min_t)
plt.plot(day,avg_t)
```

```python
# In the above graph, by seeing the we donot know which lines means what and 
# what is x-axis represent and what y-axis represents
# Label is used for the name of x and y axis, Title is used for the title of the graph
plt.xlabel("Days")
plt.ylabel("Tempreture")
plt.title("Weather")

plt.plot(day,max_t)
plt.plot(day,min_t)
plt.plot(day,avg_t)
```

```python
# Now we have legend function which helps to indentify each line of graph, label of each plotted graph line is
# requied to legend function to work
plt.xlabel("Days")
plt.ylabel("Tempreture")
plt.title("Weather")

plt.plot(day,max_t, label='Max')
plt.plot(day,min_t, label='Min')
plt.plot(day,avg_t, label='Avg')

plt.legend()
```

```python
# If legend() function as no argument, it by default takes the legend(loc='Best') argument
# sometime we want to change the position of legend by our requirement
plt.xlabel("Days")
plt.ylabel("Tempreture")
plt.title("Weather")

plt.plot(day,max_t, label='Max')
plt.plot(day,min_t, label='Min')
plt.plot(day,avg_t, label='Avg')

plt.legend(loc='upper right')
```

```python
# other properties of legend are shadow and forsize etc we can check all the properties of each fuction in the 
# available api document try out all properties on a browser search
plt.xlabel("Days")
plt.ylabel("Tempreture")
plt.title("Weather")

plt.plot(day,max_t, label='Max')
plt.plot(day,min_t, label='Min')
plt.plot(day,avg_t, label='Avg')

# plt.legend(loc='best', shadow=True, fontsize='large')
plt.legend(loc='best', shadow=True, fontsize='small')
```

```python
# Grid is used to read the graph easily 
plt.xlabel("Days")
plt.ylabel("Tempreture")
plt.title("Weather")

plt.plot(day,max_t, label='Max')
plt.plot(day,min_t, label='Min')
plt.plot(day,avg_t, label='Avg')

# plt.legend(loc='best', shadow=True, fontsize='large')
plt.legend(loc='best', shadow=True, fontsize='small')
plt.grid()
```
